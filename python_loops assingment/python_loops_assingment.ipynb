```python
# calculate the sum of the numbers from 1 to 100

a = 0
for i in range(1,101):
  a += i
print(a)
```

```python
# print multiplication table of a given number(e.g.7)

number = int(input("Enter a number: "))
a = 1
while a < 11:
  print(number,"*", a, "=", number*a)
  a += 1
```

```python
# reverse a number using a for loop
example: 1234-4321

for i in range(4,0,-1):
  print(i)

```

```python
# find the factorial number 4

number = 4
a = 1

while number > 0:
  a *= number
  number -= 1
print(a)
```

```python
# check weather a number is prime using a loop in python

a = 29

if a > 1:
  for i in range(2,a):
    if a % i == 0:
      print(a, "is not a prime number")
      break
  else:
    print(a, "is a prime number")
else:
  print(a, "is not a prime number")
```
